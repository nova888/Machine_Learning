```python
# Read Dataset
import pandas as pd
import matplotlib.pyplot as plt

```

```python
dskalori = pd.read_excel('kaloriberat.xlsx')
```

```python
dskalori.head()
```

```python
print(dskalori)
```

```python
dskalori.info()
```

```python
plt.scatter(dskalori['Kalori'],dskalori['Berat'])
plt.xlabel('Kalori')
plt.ylabel('Berat')
```

```python
#Independen dan Dependen Variable
```

```python
x = dskalori.iloc[:,[0]].values
y = dskalori.iloc[:,[1]].values
```

```python
# split trend test

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=0)
```

```python
x_test
```

```python
#model regresi

from sklearn.linear_model import LinearRegression
regresi = LinearRegression()
```

```python
regresi.fit(x_train,y_train)
```

```python
#visualisasi garis regresi

plt.scatter(x_train, y_train)
plt.plot(x_train, regresi.predict(x_train), c='red')
plt.xlabel('Kalori')
plt.ylabel('Berat')
```

```python
#Testing Model
y_prediksi = regresi.predict(x_test)
print (y_prediksi)
```

```python
#test Data Baru

data_baru = [[600]]
prediksi_baru = regresi.predict(data_baru)
print(prediksi_baru)
```
